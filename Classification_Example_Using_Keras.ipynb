```python
import keras
import os
import numpy as np
import random
import cv2
```

```python
from keras.layers import Input, Dense, BatchNormalization, Activation, Conv2D, Dropout, ReLU, Reshape
from keras.models import Sequential
from keras import Model
```

```python
Layer_1 = Input(shape=(64,64,3))
Layer_2 = Conv2D(filters=16, strides=2, kernel_size=3, padding="valid", activation="relu")(Layer_1)
Layer_3 = Dropout(rate=0.2)(Layer_2)
Layer_4 = BatchNormalization()(Layer_3)
```

```python
Layer_5 = Conv2D(filters=16, strides=2, kernel_size=3, padding="valid", activation="relu")(Layer_4)
Layer_6 = Dropout(rate=0.2)(Layer_5)
Layer_7 = BatchNormalization()(Layer_6)
```

```python
Layer_8 = Conv2D(filters=2, strides=1, kernel_size=15)(Layer_7)
Layer_9 = Reshape(target_shape=(2,))(Layer_8)
```

```python
network_model = Model(inputs=Layer_1, outputs=Layer_9)
```

```python
network_model.summary()
```

```python
network_model.compile(metrics=["accuracy"], optimizer="adam", loss="categorical_crossentropy")
```

```python
train = []
for root,sub,file in os.walk("dogs-vs-cats/train/train"):
    for file_name in file:
        train.append(root+"/"+file_name)
random.shuffle(train)
```

```python
label_dict = {"cat":[0,1], "dog":[1,0]}
for j in range(0,len(train)-32,32):
    image = []
    label = []
    for k in range(j,j+32):
        img = cv2.imread(train[k])
        img = cv2.resize(img, (64,64))
        img = img/255.0 - 0.5
        img = np.array(img).astype("float32")
        image.append(img)
        label.append(label_dict[train[k].split("/")[-1].split(".")[0]])   
    image_ = np.array(image).reshape(32,64,64,3)
    label_ = np.array(label).reshape(32,2)
    network_model.fit(image_, label_, epochs=10, batch_size=32)
#     print(loss_val)
#         break
#     break
```
